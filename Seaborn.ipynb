```python
# Import libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Get dataset names which exist in seaborn library 
sns.get_dataset_names()
```

```python
# Import datasets
tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
planets = sns.load_dataset("planets")
```

```python
planets.head()
```

```python
# Plotting number of method types which is in planets dataset
plt.figure(figsize=(12, 6), dpi=80)
plt.xticks(rotation=45)
sns.set_theme(style="darkgrid")
sns.countplot(x= planets["method"], data=planets, hue="method",dodge=False)
plt.plot()
```

```python
tips.head()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day",palette="YlGnBu", size="size")
plt.plot()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(tips["tip"], kde=True, bins=15, color="red")
plt.plot()
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(data=tips, x="day", y="total_bill", palette="YlGnBu", hue="sex")
plt.plot()
```

```python
plt.figure(figsize=(12,6), dpi=80)
sns.stripplot(data=tips, x="day", y="total_bill", palette="YlGnBu", hue="sex")
plt.plot()
```

```python
plt.figure(figsize=(12,6), dpi=80)
sns.jointplot(data=tips, x="tip", y="total_bill", kind="reg")
plt.plot()
```

```python
plt.figure(figsize=(12,6), dpi=80)
sns.jointplot(data=tips, x="tip", y="total_bill", kind="kde", shade=True, cmap="BuPu")
plt.plot()
```

```python
plt.figure(figsize=(12,6), dpi=80)
sns.jointplot(data=tips, x="tip", y="total_bill", kind="hex", cmap="YlGnBu") # There is no shade parameter here
plt.plot()
```
